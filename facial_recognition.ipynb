```python

import face_recognition
import numpy as np
import cv2
import os
from PIL import Image, ImageDraw
```

```python
KNOWN_FACES_DIR = r'/content/known_faces'
KNOWN_FACES_DIR =r'/content/known_faces' 
TOLERANCE = 0.6 
FRAME_THICKNESS = 3
FONT_THICKNESS = 2
MODEL = "cnn" #hog

print("loading known faces")

known_faces = ['known.jpg']
known_names = ['known.jpg']
for name in os.listdir(KNOWN_FACES_DIR):

  filename= os.path.join(KNOWN_FACES_DIR,name)
  if filename.endswith(".jpg"):
    image = face_recognition.load_image_file(filename)
    print(type(image))
  encoding = face_recognition.face_encodings(image)[0]
  known_faces.append(encoding)
  known_names.append(name)

print("processing known faces")

for filename in os.listdir(KNOWN_FACES_DIR):
    image = face_recognition.load_image_file(f"{KNOWN_FACES_DIR}/{filename}")
    locations = face_recognition.face_locations(image, model=MODEL)
    encodings = face_recognition.face_encodings(image, locations)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    for face_encoding, face_location in zip(encodings, locations):
        results= face_recognition.compare_faces([known_faces], face_encoding, TOLERANCE)
        results= face_recognition.compare_faces([known_faces], face_encoding)
        match = None 
        if True in results:
            match = known_names[results.index(True)]
            print(f"Match found: {match}") 
            top_left = (face_location[3], face_location[0])
            bottom_right = (face_location[1], face_location[2])
            bottom_right = np.ravel(bottom_right)
            color = [0, 255, 0]
            cv2.rectangle(image, top_left, bottom_right, color, FRAME_THICKNESS)

            top_left = (face_location[3], face_location[2])
            bottom_right = (face_location[1], face_locations[2]+22)
            cv2.rectangle(image, top_left, bottom_right, color, cv2.FILLED)
            cv2.putText(image, match, (face_location[3]+10, face_location[2]+15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (200,200,200), FONT_THICKNESS)

CV2.imshow(known, image)
cv2.waitKey(10000)
cv2.destroyWindow(unknown)
```
